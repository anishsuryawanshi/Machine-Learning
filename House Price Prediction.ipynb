### Removing Outliers

```python
cols=['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'sqft_above',
       'sqft_basement', 'rounded_prices','floors']
cols
```

```python
for i in cols:
    sns.boxplot(df[i],whis=1.5)
    plt.show()
```

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
for column in df[cols].columns:
    lr,ur=remove_outlier(df[column])
    df[column]=np.where(df[column]>ur,ur,df[column])
    df[column]=np.where(df[column]<lr,lr,df[column])
```

### Spliting Data into two parts

#### train_test_split

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

### Model Evalution 

#### we are going to import all the regressor algorithm 

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
knn=KNeighborsRegressor()
linreg=LinearRegression()
svm=SVR()
dt=DecisionTreeRegressor()
rf=RandomForestRegressor()
```

### making model to fit algorithm on data

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

knn = KNeighborsRegressor(n_neighbors=3)

def mymodel(model):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    print("Mean Squared Error:", mse)
    print("Mean Absolute Error:", mae)
    print("R-squared Score:", r2)
    
    return model
```

```python
mymodel(linreg)
```

```python
mymodel(knn)
```

```python
mymodel(dt)
```

```python
mymodel(rf)
```

```python
def house_price_prediction(model,df):
    X = df.iloc[:,:-1]
    y = df.iloc[:,-1]

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

    mymodel = model

    mymodel.fit(X_train, y_train)
    
    
    bedrooms     = int(float(input('Enter the numbers of bedrooms     : ')))
    bathrooms    = int(eval(input('Enter the bathrooms size           : ')))
    floors       = int(float(input('enter the floors                  : ')))
    sqft_living  = float(input('Enter the total square-feet of living : '))
    sqft_lot     = float(input('Enter the total square-feet of lot    : '))
    sqft_above   = float(input('Enter the total square-feet of above  : '))
    sqft_basement= float(input('Enter the total square-feet of basement : '))
    city         = int(input('enter the city code from 1 to 36        :'))
    country      = int(input('enter the country code (0)              :'))
    
    predicted_price = mymodel.predict([[bedrooms,bathrooms,city,country,sqft_living,sqft_lot,sqft_above,sqft_basement,floors]])[0]

    if predicted_price > 0:
        print(f'Predicted price for the given inputs is {predicted_price}')
    else:
        print('Price prediction not available')

    
```
